```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
from tqdm import tqdm 
import pickle 
```

```python
_DATA_SET = "D:\VARSITY\SEMESTER 10\CVPR\Assignment"
_CATEGORIES = ['airplane', 'automobile','bird','cat','deer','dog','frog','horse','ship','truck'] 
```

```python
#csv file read
csv_file = os.path.join(_DATA_SET,"trainLabels.csv")
import csv

with open(csv_file,'r') as file:
    reader = csv.reader(file)
    with open('coors_new.csv', mode='w') as outfile:
        writer = csv.writer(outfile)
        TRAIN_LABELS = {rows[0]:rows[1] for rows in reader}

print(TRAIN_LABELS)
        
      
        
```

```python
_TRAIN_IMGS=[]
IMG_SIZE = 100

path = os.path.join(_DATA_SET,"train")

for s in tqdm(TRAIN_LABELS):
    if(s=="id"):
        continue
    else:
        category = TRAIN_LABELS[s]
        img = s+".png"
        img_arr = cv2.imread(os.path.join(path, img), cv2.IMREAD_GRAYSCALE)
        img_resized = cv2.resize(img_arr, (IMG_SIZE, IMG_SIZE))
        _TRAIN_IMGS.append([img_resized,category])
       
        
    
```

```python
print(len(_TRAIN_IMGS))
```

```python
import random
random.shuffle(_TRAIN_IMGS)
```

```python
plt.figure(figsize=(14,8))
for i in range(50):
    plt.subplot(5,10, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(_TRAIN_IMGS[i][0], cmap=plt.cm.gray)
    plt.xlabel(_TRAIN_IMGS[i][1])
    
plt.show()  
```

```python
fold=_TRAIN_IMGS[0:10000]
fold1=_TRAIN_IMGS[10000:20000]
fold2=_TRAIN_IMGS[20000:30000]
fold3=_TRAIN_IMGS[30000:40000]
fold4=_TRAIN_IMGS[40000:50000]

print(len(fold))
print(len(fold1))
print(len(fold2))
print(len(fold3))
print(len(fold4))

validation_set = fold4[:]
xtarin_set = fold+fold1+fold2+fold3
print(len(xtarin_set))

```

```python
def Knn_accuracy(validation_set,xtarin_set,k):
    plt.figure(figsize=(18,18))
    tmpIndex = 1
    n=1
    checking = 50
    v=k
    valid_arr = []

    for i in range(checking): 
       
    
        valid = 0
        accuricy = []
        arr_list =[]
        for j,k in tqdm(enumerate(xtarin_set)):
            dist_arr = np.abs(validation_set[i][0] - xtarin_set[j][0]) 
            dist = np.sum(dist_arr) # mathatten / L1 distance
            arr_list.append([dist, xtarin_set[j][0], xtarin_set[j][1]])
    
        arr_list.sort(key=lambda e:e[0])
        #k = 10
        neighbors = []
    
    
        for z in range(v):
            neighbors.append([dist, xtarin_set[z][0], xtarin_set[z][1]])
    
      
    
        
        for m in tqdm(range(v)):
            
            if validation_set[i][1] == arr_list[m][2]:
                valid = valid+1
            n=n+1
            
        print("vlaid=",valid)
        valid_arr.append(valid)
        tmpIndex += 11
        n=n+1
    
    total_valid = sum(valid_arr)
    accuracy = (total_valid/(v*checking))*100
    print(accuricy)
    
    return accuracy

    
    
    
k_array=[1,2,3,4,5]
accuaracy_array = []
accuracy_avg = []


#1st fold
accuracy_count1 = []
validation_set = fold4[:]
xtarin_set = fold+fold1+fold2+fold3

for i in range(len(k_array)):
    accuracy=Knn_accuracy(validation_set,xtarin_set,k_array[i])
    print("accuracy ",i,": ",accuracy)
    accuracy_count1.append(accuracy)

accuaracy_array.append(accuracy_count1)
avg = sum(accuracy_count1)/len(accuracy_count1)
accuracy_avg.append(avg)

print(accuaracy_array)
print(accuracy_avg)



#2nd fold
accuracy_count2 = []
validation_set = fold3[:]
xtarin_set = fold+fold1+fold2+fold4

for i in range(len(k_array)):
    accuracy=Knn_accuracy(validation_set,xtarin_set,k_array[i])
    print("accuracy ",i,": ",accuracy)
    accuracy_count2.append(accuracy)

accuaracy_array.append(accuracy_count2)
avg = sum(accuracy_count2)/len(accuracy_count2)
accuracy_avg.append(avg)

print(accuaracy_array)
print(accuracy_avg)


#3rd fold
accuracy_count3 = []
validation_set = fold[:]
xtarin_set = fold1+fold2+fold3+fold4

for i in range(len(k_array)):
    accuracy=Knn_accuracy(validation_set,xtarin_set,k_array[i])
    print("accuracy ",i,": ",accuracy)
    accuracy_count3.append(accuracy)

accuaracy_array.append(accuracy_count3)
avg = sum(accuracy_count3)/len(accuracy_count3)
accuracy_avg.append(avg)

print(accuaracy_array)
print(accuracy_avg)




#4th fold
accuracy_count4 = []
validation_set = fold1[:]
xtarin_set = fold+fold2+fold3+fold4

for i in range(len(k_array)):
    accuracy=Knn_accuracy(validation_set,xtarin_set,k_array[i])
    print("accuracy ",i,": ",accuracy)
    accuracy_count4.append(accuracy)

accuaracy_array.append(accuracy_count4)
avg = sum(accuracy_count4)/len(accuracy_count4)
accuracy_avg.append(avg)

print(accuaracy_array)
print(accuracy_avg)



#5th fold
accuracy_count5 = []
validation_set = fold[:]
xtarin_set = fold1+fold2+fold3+fold4

for i in range(len(k_array)):
    accuracy=Knn_accuracy(validation_set,xtarin_set,k_array[i])
    print("accuracy ",i,": ",accuracy)
    accuracy_count5.append(accuracy)

accuaracy_array.append(accuracy_count5)
avg = sum(accuracy_count5)/len(accuracy_count5)
accuracy_avg.append(avg)

print(accuaracy_array)
print(accuracy_avg)








```
